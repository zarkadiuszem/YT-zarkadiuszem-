#Pandas
Pakiet do przetwarzania danych. <br>
Oferuje strukturę danych **data frame**, która przypomina arkusz np. Excel.<br>
Tak jak w arkuszu mamy tu nazwy (identyfikatory, etykiety) kolumn i wierszy, co jest typowe dla analizy danych empirycznych. <br>
W przypadku NumPy (tablice) takich etykiet nie mieliśmy.

```python
import pandas as pd
import numpy as np
```

Podstawową strukturą jest ramka danych ***data frame***, ale zaczniemy od prostszej struktury ***series***

##Series (szereg, szereg szczegółowy)
Jednowymiarowa struktura danych z indeksami

```python
pd.Series(data=[1,2,3,4,5]) # korzystamy tu z klasy Series aby stworzyć obiekt Series
#np.array([1,2,3,4,5])
```

```python
df = pd.Series(data=[1,2,3,4,5])
print(df)
df = df.drop(1) # metoda drop (python)
df
```

```python
df.iloc[1] # nr wiersza (metody dostępu iloc i loc)
```

```python
df.loc[2] # nazwa etykiety
```

```python
df = df + 2 # wektoryzacja
df
```

```python
df = pd.Series(data=[1,2,3,4,5], index=['arek','marek','darek','jarek','arek'])
df
```

```python
df['arek'] # indeksy/etykiety nie muszą być unikalne (uzyskujmy ponownie series)
```

```python
df.loc['arek'] # to samo
```

```python
df.iloc[4]
```

```python
x = df.loc['arek'].iloc[1]
x + x
```

```python
x = int(x)
x
```

```python
sum(df)
```

```python
np.std(df['arek'])
```

```python
df.describe() # metoda w Pandas
```

```python
df.describe(percentiles=[0.1,0.5,0.9])
```

```python
df.info()
```

##DataFrame v1

###Tworzenie DataFrame

```python
df = pd.DataFrame(data = [[1,2],[3,4],[5,6]],   # korzystamy z klasy DataFrame
             index = ['arek','darek','marek'],  # dwuwymiarowa struktura danych - arkusz
             columns = ['zł','$'])
df
```

```python
df.info()
```

```python
df['zł'] *= .1 # wektoryzacja
df
```

### Wybór wierszy i kolumn

```python
print(df['zł']) # wybór kolumny
type(df['zł'])
```

```python
df.zł   # o ile nie ma nazw kolumn ze spacjami np. "dochody zł"
```

```python
# skoro df['zł'] zwraca Series to za pomocą drugiego nawiasu możemy wskazać wiersz/e
df['zł']['arek'] # kolumna - wiersz (separowane nawiasami)
```

```python
df['zł'][['arek','darek']]
```

```python
np.std(df['zł'])
```

```python
df.describe() # prosty opis statystyczny
```

```python
df['zł'].describe() # opis jednej zmiennej (ilościowej)
```

```python
print(df.columns)
print(df.index)
```

```python
df['razem_zł'] = df['zł'] + df['$']*4 # działania na kolumnach
df
```

### Selektory *loc i iloc*

```python
df.loc['arek']
```

```python
df.loc['arek','zł'] # loc i iloc wiersz przecinek kolumna (separowane przecinkami)
```

```python
df.loc[['arek','darek'],'zł']
```

```python
df.iloc[0:3, 0]
```

```python
df.loc['arek', '$']
# df.iloc[0,1] to samo
```

### Usuwanie kolumn/wierszy

```python
df.drop('zł', axis = 1) # domyślnie inplace = False = nie zmieniamy df
# jeśli chcemy zmienić to albo inplace = True albo df = df.drop('zł', axis = 1)
# df.drop('zł', axis = 1, inplace = True)
df
```

```python
df.drop('arek', axis = 0)
```

```python
print(df.drop(['arek','marek'], axis = 0))
print(df.drop(index = ['arek','marek']))
print(df.drop(columns = ['zł','$']))
```
